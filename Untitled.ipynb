```python
import matplotlib.pyplot as plt
import pandas as pd
from get_all_madison_meetups import get_madison_tech_groups, get_group_events

%matplotlib inline
```

```python
madison_tech = get_madison_tech_groups()
tech_data = pd.DataFrame.from_records(madison_tech)
```

```python
tech_data.columns
```

```python
membership_df = tech_data.sort_values('members', ascending=True)[['name', 'members']]
membership_df.head()
```

```python
fig, ax = plt.subplots(figsize=(20, 15))
membership_df.set_index('name').plot(kind='barh', ax=ax)
ax.set_xlabel('Number of members')
ax.set_ylabel('Meetup group')
plt.show()
```

```python
group_events = tech_data.id.apply(get_group_events)
```

```python
group_events2 = pd.concat(list(group_events), ignore_index=True)
```

```python
group_events2.group[0]
```

```python
group_events2['name'] = group_events2.group.apply(lambda d: d['name'])
```

```python
fig, ax = plt.subplots(figsize=(20, 15))
group_events2.groupby('name').size().plot(kind='barh', ax=ax)
ax.set_xlabel('Cumulative number of events')
ax.set_ylabel('Meetup group')
plt.show()
```
